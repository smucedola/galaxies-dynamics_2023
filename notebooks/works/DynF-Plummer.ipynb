## Massive perturber in a plummer sphere
We aim to initialize a massive perturber inside a plummer sphere, with initial velocity given by the equilibrium condition and analyze its evolution.

First let's start by initializing a plummer sphere.

## Initial positions

The Plummer sphere density profile is

$$
\rho \left( R \right) = \frac{3 M_{tot}}{4\pi b^3}\  \left(1 + \dfrac{r^2}{b^2}\right)^{-5/2}
$$

$\newline$

To initialize the Plummer sphere we need to set the correct distributions for our particles cartesian coordinates. To do so we first need to find the spherical coordinates and then project them onto the cartesian plane. 

The mass can be found by simple integration over the volume, which yields

$$
    M(r) = M_\text{tot}\ \dfrac{r^3}{b^3}\ \left(1 + \dfrac{r^2}{b^2}\right)^{-3/2}
$$
$\newline$

Inverting the relation for $\xi = M(r)/M_{tot}$ we obtain that 

$$
\begin{cases}
    \ R\ =\ b\ \sqrt{\left( \dfrac{ \xi^{2/3}}{1-\xi^{2/3}}\right)} && 0 \le \xi \le 1 \\ \\
    \ \varphi = \tilde{\varphi} && 0 \le \tilde{\varphi} \le 2 \pi \\ \\
    \ \theta = \arccos(1-2 \omega) &&  0 \le \omega \le 1 
\end{cases}
$$

$\newline$

Now that we have the spherical coordinates we need to initialize their distribution, in this case a direct montecarlo takes care of it. We just initialize the cumulative and calculate the radii

Finally we calculate the potential, which is simply 

$$
    \phi = - \dfrac{GM_\text{tot}}{\sqrt{r^2 + b^2}}
$$

But in the new formalism we will use the following quantity 

$$
    \psi = - \phi + \phi_0 = \dfrac{GM_\text{tot}}{\sqrt{r^2 + b^2}}
$$

```python
# we are still in internal units
G = 1
M = np.sum(sphere.m)
R = sphere.r

Psi = G * M /(R**2 + b**2)**(1/2)
```

## Initial velocities

Then we can initialize the velocities for the equlibrium condition. The distribution function is 

$$f(x) = A q^2 \ \left( 1 - q^2 \right)^{7/2}$$

where
$$
\begin{cases}
   \vert \vec{v} \vert = q \sqrt{2\Psi} &&  0 \le q \le 1  && \text{with rejection on $F(q)$} \\ \\
    v_{\varphi} = \tilde{\varphi} && 0 \le \tilde{\varphi} \le 2 \pi \\ \\
    v_{\theta} = \arccos(1-2 \omega) &&  0 \le \omega \le 1 
\end{cases}
$$

Let's sample the values of $q$ using rejection method, first we estimate A

```python
def f_not_normalized(q):
    return q**2*(1-q**2)**(7/2)
```

```python
from scipy.integrate import quad

q_ = np.linspace(0,1,N*100) #more points=more accurate integration
g = np.max(f_not_normalized(q_))

int_ = quad(f_not_normalized,min(q_), max(q_))[0] # computing the normalization

A = 1/int_ #normalizing f
```

```python
def f(q, A):
    return A*q**2*(1-q**2)**(7/2)
```

```python
g = np.max(f(q_, A))

integral = quad(f, min(q_), max(q_), args=(A))[0] #checking normalization now

plt.plot(q_, f(q_, A), label= fr'Integral = {integral :.2f}') # must be one
plt.scatter(q_[g==f(q_, A)], g, c='r', s=10, zorder=2, 
            label=fr'$g = \mathrm{{max}}(f) = {g:.2f}$')

plt.title(r'$q$ pdf')
plt.legend()

plt.show()
```

```python
def rejection_sampling(f, g, num_samples):
    samples = []
    
    with tqdm(total=num_samples, desc='Sampling') as pbar:
        while len(samples) < num_samples:
            
            q_i = np.random.uniform(0, 1)
            u = np.random.uniform(0, 1)
            
            # Accept or reject the candidate
            if u < f(q_i, A) / g: 
                samples.append(q_i)
                pbar.update(1)
                
    return np.array(samples)
```

```python
num_samples = N

q_sample = rejection_sampling(f,g, num_samples)
```

```python
plt.hist(q_sample, bins=50, alpha=.3, density=True)
plt.plot(q_, f(q_, A), label= fr'Integral = {integral :.2f}', lw=1)
plt.scatter(q_[g==f(q_, A)], g, c='r', s=10, zorder=2, 
            label=fr'$g = \mathrm{{max}}(f) = {g:.2f}$')

plt.title(r'$q$ pdf')
plt.legend()

plt.show()
```

Once we computed the samples through the Monte-Carlo rejection method, we can set up the velocities for each particle, using

```python
v = q_sample * np.sqrt(2 * Psi)
v_theta_cdf = np.random.uniform(0,1,N)
v_theta = np.arccos(1 - 2 * v_theta_cdf)
v_phi = np.random.uniform(0, 2 * np.pi, size=N)

vx = v * np.cos(v_phi) * np.sin(v_theta)
vy = v * np.sin(v_phi) * np.sin(v_theta)
vz = v * np.cos(v_theta)
```

```python
sphere.set_velocities(vx, vy, vz)
```

```python
sphere.projections(s=.5,a=.1, l=50);
```

```python
sphere.v_projections(s=1, a=.1, l=0.5);
```

## Adding the perturber
We have initialized a plummer sphere of 5000 stars, now let's add the perturber. The effect is maximum if $M_\mathrm{pert} \gg m_*$ and $v_\mathrm{pert} \sim v_m$

```python
M_p = 1e2*sphere.m[0]
R_p = 1.5*b

#choosing the initial coordinates
x_p = R_p
y_p = 0
z_p = 0
```

For the velocities we set a circular orbit for the perturber, we know that

$$ v_\mathrm{circ} = \sqrt{\dfrac{GM(R)}{R}} $$

therefore 

$$ v_\mathrm{circ} = \sqrt{\dfrac{G M_\text{tot}\ \dfrac{R^3}{b^3}\ \left(1 + \dfrac{R^2}{b^2}\right)^{-3/2}}{R}} $$

```python
vx_p = 0 
vy_p = np.sqrt(G*M*(R_p**2/b**3)*(1+(R_p/b)**2)**(-3/2))
vz_p = 0
```

Now let's append these data to the sphere positions and velocities vectors

```python
properties = ['m', 'r', 'x', 'y', 'z', 'vx', 'vy', 'vz']
values = [M_p, R_p, x_p, y_p, z_p, vx_p, vy_p, vz_p]

for prop, val in zip(properties, values):
    setattr(sphere, prop, np.append(getattr(sphere, prop), val)) #adding the perturber
```

## Sending the simulation

```python
home = '../../'
input = home + f'data/inputs/plummer_dyf{N}.txt' 
output = home + f'data/outputs/plummer_dyf{N}.txt'
log = f'../../data/outputs/plummer_dyf{N}_log.txt'
```

```python
open(input, 'w').close()  #clearing input file
open(output, 'w').close() #clearing output file
open(log, 'w').close()    #clearing log file

sphere.write_tree(input)
```

Now we calculate the softening and accuracy parameters

```python
V = 4/3*np.pi*b**3
epsilon = 1/100 * (V/int(float(N)))**(1/3)
T = np.sqrt(3 * np.pi/ (32*sphere.rho_mean_sphere(b))) #dynamical time
```

Instead of initializing the time as a function of the dynamical time, we wish to determine the secular timescale from the theoretical derivation. Unfortunately, the plummer case is not as easy as the singular isothermal sphere so we will instead do it by hand

```python
T_sim = round(20 * T)
dt = 1/4
theta = 0.1
out = 4*dt

print(f'T_sim = {T_sim:.1f} \t epsilon = {epsilon:.2e} \t dt = {dt:.2f} \t out = {out:.2f}')
```

```python
os.system(home + f'treecode/tree/treecode in={input} out={output} eps={epsilon} tstop={T_sim} dtime={dt} dtout={out}>{log}')
```
