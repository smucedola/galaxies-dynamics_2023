```python
import matplotlib
import numpy as np
import random
import pandas as pd
import scipy.integrate as integrate
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=False)
```

```python
M = 1
N = 10**3
b = 10
m = M/N
G = 1
```

```python
r_cdf = np.random.uniform(0,1,N)
R = b * np.sqrt(r_cdf**(2/3)/(1- r_cdf**(2/3)))
theta_cdf = np.random.uniform(0,1,N)
theta = np.arccos(1 - 2 * theta_cdf)
phi = np.random.uniform(0, 2 * np.pi, size=N) 
```

```python
fig, axs = plt.subplots(1,3, tight_layout=True, sharey=True)
axs[0].set(title = fr'$r$ distribution', xlabel = r'$r$', ylabel = r'$p(r)$')
axs[0].hist(R, bins = 30, density = True, alpha = .5, color='royalblue', lw = 0)
axs[1].set(title = fr'$\theta$ distribution', xlabel = r'$\theta$', ylabel = r'$p(\theta)$' )
axs[1].hist(theta, bins = 30, density = True, alpha = .5, color='royalblue', lw = 0)
axs[2].set(title = fr'$\phi$ distribution', xlabel = r'$\phi$', ylabel = r'$p(\phi)$' )
axs[2].hist(phi, bins = 30, density = True, alpha = .5, color='royalblue', lw = 0);
```

```python
x = R * np.cos(phi) * np.sin(theta)
y = R * np.sin(phi) * np.sin(theta)
z = R * np.cos(theta)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

l = 100
ax.set_xlim(-l, l)
ax.set_ylim(-l, l)
ax.set_zlim(-l, l)
ax.scatter(x, y, z,alpha = .1, s=.5)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show();
```

```python
Psi = G * M /(R**2 + b**2)**(1/2)
```

```python
pdf_test = lambda x: (x**2*(1-x**2)**(3.5))
A = integrate.quad(pdf_test,0,1)
q = np.linspace(0,1,100000)
pdf = 1/A[0]*(q**2*(1-q**2)**(3.5))
pdf_func = lambda x: 1/A[0]*(x**2*(1-x**2)**(3.5))
g = np.ones(100000)*np.max(pdf)
```

```python
q_sample = []

for i in range(N):
    y_rand = np.max(g) + 1
    rand = np.random.uniform(0,1)
    
    while y_rand > pdf_func(rand):
        rand = np.random.uniform(0,1)
        y_rand = np.random.uniform(0,np.max(g))

    q_sample.append(rand)
```

```python
q_sample = np.array(q_sample)
plt.hist(q_sample,100, density=True, histtype='stepfilled', alpha = .3)
plt.plot(q,pdf);
```

```python
v = q_sample * np.sqrt(2 * Psi)
v_theta_cdf = np.random.uniform(0,1,N)
v_theta = np.arccos(1 - 2 * v_theta_cdf)
v_phi = np.random.uniform(0, 2 * np.pi, size=N) 
```

```python
v_x = v * np.cos(v_phi) * np.sin(v_theta)
v_y = v * np.sin(v_phi) * np.sin(v_theta)
v_z = v * np.cos(v_theta)
```

```python
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(projection='3d')

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)
ax.scatter(v_x, v_y, v_z, s=.5, alpha = 0.05)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show();
```

### Adding the perturber

```python
# creating the input file
m_arr = m * np.ones(N)
dim = 3
```

```python
M_p = 0.03
R_p = 10
raggi_i = (x**2+y**2+z**2)**0.5
provv = len([el for el in (raggi_i) if (el<R_p)])
M_inside = provv * m
v_p   = ((M_inside*R_p**2)/((R_p**2+b**2)**(3/2)))**(1/2)
m_arr = np.insert(m_arr,0,M_p)
x = np.insert(x,0,R_p)
y = np.insert(y,0,0)
z = np.insert(z,0,0)
v_x = np.insert(v_x,0,0)
v_y = np.insert(v_y,0,v_p)
v_z = np.insert(v_z,0,0)
```

```python
print(len(x))
```

```python
N += 1
```

```python
with open ('in_data.txt', 'w+') as f:
    f.write(str(N) + '\n')
    f.write(f'{dim}' + '\n')
    f.write('0' + '\n')
    for m_i in m_arr:
        f.write(f'{m_i}' + '\n')
    for i in range(N):
        f.write(f'{x[i]} {y[i]} {z[i]}' + '\n')
    for j in range(N):
        f.write(f'{v_x[j]} {v_y[j]} {v_z[j]}' + '\n')
```

```python
# simulation
from matplotlib import animation
from tqdm.notebook import tqdm

t_sim = t_s[::5]

def update(num):
    global N    #N_particles
    l = 100
    ax.clear()
    
    ax.scatter(x_s[0,num-1], y_s[0,num-1], z_s[0,num-1], color = 'red', s = 5 ,alpha = .3) # perturber position
    ax.scatter(x_s[1:,num-1], y_s[1:,num-1], z_s[1:,num-1], color = 'royalblue', s = .5 ,alpha = .3) # Updating Point Location

    ax.set_title(f'{N} particles, mass = {round(m_s[0,0], 5)}, time = {round(t_sim[num],1)}')
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$y$')
    ax.set_zlabel(r'$z$')
    
    ax.set_xlim(-l, l)
    ax.set_ylim(-l, l)
    ax.set_zlim(-l, l)
    

## Plotting the Animation
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set(aspect='equal')
line_ani = animation.FuncAnimation(fig, update, interval=1,   
                                   frames=tqdm(range(len(t_sim) - 1)))

# Saving the Animation
f = r"plummer.gif"
writergif = animation.PillowWriter(fps=30)
line_ani.save(f, writer=writergif, dpi=150)

plt.show()
```
